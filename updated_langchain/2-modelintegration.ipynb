```python
import os
from dotenv import load_dotenv
load_dotenv()

GROQ_API_KEY = os.getenv("GROQ_API_KEY")
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
response=model.invoke("why do parrots talk?")
response.content
```

Streaming and batch


```python
response=model.invoke("write me a 200 words paragraph on Artificial Intelligence")
response.content
```

```python
for chunk in model.stream("write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text, end="|", flush=True)
```

```python
responses=model.batch([
    "why do parrots have colorful feathers?",
    "how do ariplane fly?",
    "when did tree stops growing anf start decaying?"
])
for response in responses:
    print(response.content)
```

```python
responses=model.batch([
    "why do parrots have colorful feathers?",
    "how do ariplane fly?",
    "when did tree stops growing anf start decaying?"
], 
config={
    "max_concurrency":5,
})
for response in responses:
    print(response.content)
```
